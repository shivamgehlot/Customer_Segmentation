```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from datetime import datetime as dt, timedelta
from sklearn.preprocessing import StandardScaler
```

```python
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
```

```python
#it give today's date use this in the real world application 
reference_date = pd.Timestamp(dt.now().date())
```

```python
# but for now we are using maximum date of dataframe  plus one day because the data is old 
reference_date = df['InvoiceDate'].max() + timedelta(days = 1)
```

```python
reference_date
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate' : lambda x : (reference_date - x.max()).days,
    'InvoiceNo' : 'nunique',
    'TotalAmount' : 'sum'
})
```

```python
rfm.head()
```

```python
rfm.rename(columns={'InvoiceDate':'Recency', 'InvoiceNo':'Frequency', 'TotalAmount':'Monetary'}, inplace=True)
```

```python
print(rfm.shape)
```

```python
rfm.head()
```

```python
scaler =  StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency','Frequency','Monetary']])
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state = 42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(8,4))
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5,init = 'k-means++',  random_state = 42)
kmeans.fit(rfm_scaled)
```

```python
labels = kmeans.labels_
```

```python
labels.max()
```

```python
labels.min()
```

```python
rfm
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
#fit only fit the data but fit_transform converts the data to array
principal_components = pca.fit_transform(rfm)
principal_components
```

```python
pca_df = pd.DataFrame(principal_components,columns=['PCA1','PCA2'])
pca_df.head()
```

```python
#concatenate the clusters labels with pca dataframe
pca_df = pd.concat([pca_df,pd.DataFrame({'Cluster Labels' : labels})],axis = 1)
pca_df.head()
```

```python
sns.scatterplot(x="PCA1",y='PCA2',hue = "Cluster Labels" , data = pca_df, palette=["red","green","blue","magenta","yellow"])
```
